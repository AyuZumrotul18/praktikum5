```python
#import library
import cv2 #library pengolahan gambar dan video 
import imageio #untuk membaca dan menulis gambar dan video dalam berbagai format.
import numpy as np #untuk mengolah array dan operasi numerik pada Python.
import matplotlib.pyplot as plt #untuk membuat visualisasi grafis dari data, seperti grafik dan plot.
```

```python
#Masukkan gambar
img = imageio.imread("minion.jpg") # untuk membaca gambar dalam berbagai format, seperti JPEG, PNG, dan lain-lain.
plt.imshow(img) # untuk menampilkan gambar dalam bentuk plot
plt.show() # untuk menampilkan plot yang dihasilkan
```

```python
#define reso dan tipe gambar
img_height= img.shape[0] # untuk mengambil ketinggian (jumlah piksel vertikal) gambar yang disimpan dalam variabel
img_width= img.shape[1] #untuk mengambil lebar (jumlah piksel horizontal) gambar yang disimpan dalam variabel
img_channel= img.shape[2] #untuk mengambil jumlah saluran (channel) gambar yang disimpan dalam variabel
img_type = img.dtype # untuk mengambil tipe data (data type) gambar yang disimpan dalam variabel 
```

```python
#membuat brightness untuk gambar grayscale
#membuat variable img_brightness untuk menampung
img_brightness = np.zeros(img.shape, dtype=np.uint8) # untuk membuat array kosong dengan ukuran dan tipe data yang sama dengan gambar yang disimpan dalam variabel 
```

```python
#membuat fungsi penambahan brightness dengan nilai yang menjadi parameter
def brighter(nilai): #deklarasi fungsi yang memiliki satu parameter
    for y  in range(0, img_height): # looping yang digunakan untuk mengakses setiap baris gambar dengan variabel y sebagai indeks baris.
        for x in range(0, img_width): #adalah looping yang digunakan untuk mengakses setiap kolom gambar dengan variabel x sebagai indeks kolom.
            red= img[y][x][0] # adalah sintaks untuk mengambil nilai komponen warna merah, hijau, dan biru pada piksel yang sedang diproses
            green= img[y][x][1]
            blue= img[y][x][2]
            gray= (int(red)+ int(green)+int(blue))/3 #sintaks untuk menghitung nilai kecerahan piksel
            gray += nilai #sintaks untuk menghitung nilai kecerahan piksel ke setiap nilai kecerahan piksel.
            if gray > 255: #kondisi jika nilai kecerahan piksel lebih besar dari 255, maka akan diberi nilai maksimum 255.
                gray = 255 
            if gray<0:
                gray = 0
            img_brightness[y][x] = (gray, gray, gray) #untuk menetapkan nilai kecerahan yang telah dihitung dan dibatasi pada piksel
```

```python
#menampilkan gambar dengan parameter -100 dan +100
fig, axes = plt.subplots(1, 2, figsize=(8, 8)) #untuk membuat plot dengan satu baris dan dua kolom 
ax = axes.ravel() # untuk membuat array 1 dimensi dari objek 

brighter(-100) #untuk menjalankan fungsi brighter() dengan parameter -100
ax[0].imshow(img_brightness) # untuk menampilkan gambar dengan kecerahan
ax[0].set_title("Brightness -100")  #untuk menetapkan judul "Brightness -100"

brighter(100) #untuk menjalankan fungsi brighter() dengan parameter 100
ax[1].imshow(img_brightness) #untuk menampilkan gambar dengan kecerahan 
ax[1].set_title("Brightness +100")  #untuk menetapkan judul "Brightness +100" 

fig.tight_layout #untuk menyesuaikan tata letak plot agar terlihat rapi.
plt.show() #untuk menampilkan plot pada layar.
```

```python
#Brightness RGB
img_rgbbright = np.zeros(img.shape, dtype=np.uint8)
```

```python
#fungsi untuk brightness RGB dengan nilai parameter 
def rgbbright(nilai):
    for y in range (0, img_height):
        for x in range (0, img_width):
            red= img[y][x][0]
            red+= nilai
            if red > 255:
                red = 255
            if red < 0:
                red = 0
            green= img[y][x][1]
            green+= nilai
            if green > 255:
                green = 255
            if green < 0:
                green = 0
            blue= img[y][x][2]
            blue+= nilai
            if blue > 255:
                blue = 255
            if blue < 0:
                blue = 0
            img_rgbbright[y][x] = (red, green, blue)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(8, 8)) 
ax = axes.ravel() 

rgbbright(-100)
ax[0].imshow(img_rgbbright) 
ax[0].set_title("Brightness -100") 

rgbbright(100)
ax[1].imshow(img_rgbbright) 
ax[1].set_title("Brightness +100") 

fig.tight_layout
plt.show()
```

```python
#contrast
img_contrass = np.zeros(img.shape, dtype=np.uint8)
```

```python
def contrass(nilai):
    for y in range (0, img_height):
        for x in range (0, img_width):
            red = img[y][x][0]
            green = img[y][x][1]
            blue = img[y][x][2]
            gray= (int(red)+ int(green)+int(blue))/3
            gray+= nilai
            if gray > 255:
                gray = 255
            img_contrass[y][x] = (gray, gray, gray) 
```

```python
fig, axes = plt.subplots(1, 2, figsize=(8, 8)) 
ax = axes.ravel() 

contrass(2)
ax[0].imshow(img_contrass) 
ax[0].set_title("Contrass 2") 

contrass(200)
ax[1].imshow(img_contrass) 
ax[1].set_title("Contrass 200") 

fig.tight_layout
plt.show()
```

```python
img_autocontrass = np.zeros(img.shape, dtype=np.uint8)
```

```python
def autocontrass():
    xmax = 300
    xmin = 0
    d = 0
    # Mendapatkan nilai d, dimana nilai d ini akan berpengaruh pada hitungan
    # untuk mendapatkan tingkat kontras
    for y in range(0, img_height):
        for x in range(0, img_width):
            red = img[y][x][0]
            green = img[y][x][1]
            blue = img[y][x][2]
            gray = (int(red) + int(green) + int(blue)) / 3
            if gray < xmax:
                xmax = gray
            if gray > xmin:
                xmin = gray
    d = xmin-xmax
    for y in range(0, img_height):
        for x in range(0, img_width):
            red = img[y][x][0]
            green = img[y][x][1]
            blue = img[y][x][2]
            gray = (int(red) + int(green) + int(blue)) / 3
            gray = int(float(255/d) * (gray-xmax))
            img_autocontrass[y][x] = (gray, gray, gray)
```

```python
autocontrass()
plt.imshow(img_autocontrass)
plt.title("Contrass Autolevel")
plt.show()
```
